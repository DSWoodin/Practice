```python
import pandas as pd
```

```python
dia = pd.read_csv('diamonds.csv')
dia.head(2)
```

```python
pd.crosstab(dia['cut'], dia['clarity'])
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=True)
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=True).round(2)
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=0) # row  
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=1) # col  
```

```python
pd.crosstab(dia['cut'], dia['clarity'], values=dia['price'], aggfunc=pd.Series.mean)
```

```python
# pd.crosstab(dia['cut'], dia['clarity'], values=dia['price'], aggfunc=pd.Series.mean)
dia.groupby(['cut', 'clarity'])['price'].mean().reset_index()
```

```python
dia_agg = dia.groupby(['cut', 'clarity'])['price'].mean().reset_index()
dia_agg.head()
```

```python
dia_agg.sort_values('price').head()
```

```python
dia_agg.sort_values('price', ascending=False).head()
```

```python
dia_agg.sort_values(['cut','clarity'], ascending=[True,False]).head()
```

```python
elec = pd.read_csv('elec_load_2017_7d.csv')
elec.head(2)
```

```python
# melt(): wide form  long form  
elec_melt = elec.melt(id_vars=['YEAR', 'MONTH', 'DAY'])
elec_melt.head(2)
```

```python
print(len(elec))
print(len(elec_melt))
```

```python
# pivot(): long form  wide form  
elec_melt.head()
```

```python
elec_melt.pivot(index=['YEAR', 'MONTH', 'DAY'],
                columns='variable', values='value')
```

```python
elec_pivot = elec_melt.pivot(index=['YEAR', 'MONTH', 'DAY'],
                columns='variable', values='value').reset_index()
elec_pivot
```

###  3:       1        ?

```python
dia = pd.read_csv('diamonds.csv')
dia.head()
```

```python
dia_agg = dia.groupby(['cut', 'color'])[['price', 'carat']].mean().reset_index()
dia_agg.head(2)
```

```python
dia_agg['ratio'] = dia_agg['price'] / dia_agg['carat']
dia_agg = dia_agg.sort_values('ratio', ascending=False)
dia_agg.head()
```
