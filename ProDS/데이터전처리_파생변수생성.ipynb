```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df['is_setosa'] = np.where(df['Species']=='setosa', 1, 0)
df.head()
```

```python
# np.where를 대체하는 방법
# True, False에 산술연산을 적용
# 0을 더하는 방법과 1을 곱하는 방법 -> 계산 전과 계산 후의 값이 같음
# 값은 그대로 유지하면서 산술연산을 적용해서 불리언변수를 수치형변수로 변환
df['is_setosa_2'] = (df['Species']=='setosa') + 0
df['is_setosa_3'] = (df['Species']=='setosa') * 1
df.head()
```

```python
bike = pd.read_csv('bike.csv')
bike.head()
```

```python
abs(bike['temp'] - bike['atemp']).mean()
```

```python
bike['diff'] = bike['temp'] - bike['atemp']
bike['diff'].abs().mean()
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike.head(2)
```

```python
bike['date'] = bike['datetime'].dt.date
bike.head(2)
```

```python
bike_agg = bike.groupby('date')['casual'].max()
bike_agg.head()
```

```python
bike_agg_up25 = bike_agg[bike_agg > 25]
bike_agg_up25.head(2)
```

```python
len(bike_agg_up25)
```

```python
# 날짜가 인덱스로 들어가서 Series가 된것이 불편하면
# reset_index()를 이용해 인덱스를 추가하면서 자연스레 DataFrame으로 변환
bike_agg = bike.groupby('date')['casual'].max().reset_index()
bike_agg.head()
```

```python
bike['hour'] = bike['datetime'].dt.hour
bike.head(2)
```

```python
bike_hour = bike.groupby('hour')['registered'].mean().reset_index()
bike_hour.head()
```
