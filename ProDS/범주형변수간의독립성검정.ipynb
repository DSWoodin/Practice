```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('financial_info_10k_persons.csv')
df.head(2)
```

```python
chi2_contingency(pd.crosstab(df['Gender'], df['Marital_status']))
```

### 문제 1: 고객의 학력이 고졸인 경우 성별과 이탈여부를 사용하여 독립성 검정을 실시했을 때 p-value는?

```python
df = pd.read_csv('financial_info_10k_persons.csv')
df.head(2)
```

```python
df_sub = df.loc[df['Edu_level'] == 'High School', ]
chi2_contingency(pd.crosstab(df_sub['Gender'], df_sub['is_attrited']),
                 correction=False)
```

### 문제 3: 최근 12개월의 이용 실적 중 3개월 이상 사용 실적이 없는 것과 고객 이탈의 관계가 있는지 독립성 검정을 실시한 결과로 옳은 것은?

```python
df = pd.read_csv('financial_info_10k_persons.csv')
df.head(2)
```

```python
df['is_inactive_last_12m'] = (df['Inactive_last_12m'] >= 3) + 0
```

```python
chi2_contingency(pd.crosstab(df['is_inactive_last_12m'], df['is_attrited']))
```
