### 문제 1: 당뇨병 발병 여부를 예측하기 위하여 의사결정나무를 사용하고자 한다. 이 때 혈당, 혈압, 임신 횟수를 기반으로 예측을 했을 때 예측 정확도는?

```python
df = pd.read_csv('diabetes.csv')
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=123)
```

```python
model = DecisionTreeClassifier(random_state=123)
model = model.fit(X = df_train.loc[:, ['Glucose','BloodPressure','Pregnancies']],
                  y = df_train['Outcome'])
pred = model.predict(df_test.loc[:, ['Glucose','BloodPressure','Pregnancies']])
pred[:4]
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true = df_test['Outcome'], y_pred = pred)
```

### 문제 2: 환자의 BMI를 예측하기 위하여 회귀나무를 사용하고자 한다. 이 때 혈당, 혈압, 피부 두께를 독립변수로 했을 경우 RMSE는 얼마인가?

```python
df = pd.read_csv('diabetes.csv')
df.head(2)
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=123)
```

```python
model = DecisionTreeRegressor(random_state=123)
model = model.fit(X = df_train.loc[:, ['Glucose','BloodPressure','SkinThickness']],
                  y = df_train['BMI'])
pred = model.predict(df_test.loc[:, ['Glucose','BloodPressure','SkinThickness']])
pred[:4]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_true = df_test['BMI'], y_pred = pred)**0.5
```

### 문제 3: 분류나무의 파라미터를 바꿔가면서 성능 평가를 하려고 한다. 당뇨 발병 여부를 종속변수로 하고 혈당, 혈압, 임신 횟수, BMI, 나이를 독립변수로 하고 Depth를 3에서 6까지 변화시킬 때 그 결과로 틀린 것은?

```python
df = pd.read_csv('diabetes.csv')
df.head(2)
```

```python
df_train, df_test = train_test_split(df, test_size=0.3, random_state=345)
```

```python
cols = ['Glucose','BloodPressure','Pregnancies','BMI','Age']
depth_list = [3,4,5,6]
accs = []
for depth in depth_list:
    model = DecisionTreeClassifier(max_depth=depth, random_state=345)
    model = model.fit(X = df_train.loc[:, cols], y = df_train['Outcome'])
    pred = model.predict(df_test.loc[:, cols])
    accs = accs + [accuracy_score(y_true = df_test['Outcome'], y_pred = pred)]
df_score = pd.DataFrame({'depth': depth_list, 'accs': accs})
df_score['accs'] = df_score['accs'].round(2)
df_score
```
