```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', \
                 dtype={'ISBN':str, '세트 ISBN':str, '주제분류번호':str})
# 판다스는 read 할때 칼럼마다 데이터 타입을 알아서 찾음 -> dtype을 직접 지정
# low_memory를 사용하면서 dtypeWarning도 발생시키지 않음
```

```python
df.head()
# csv의 첫 번째 행은 열 이름으로 인식
# 만약 첫 행이 열 이름이 아니면 read_csv()에서 header를 None, names에 열 이름 리스트를 따로 전달
# Unnamed: 13 열은 아까 csv파일을 열었을 때 각 라인 끝에 콤마(,)로 끝나서 열이 하나 더 있다고 판단했기 때문
```

```python
df.to_csv('ns_202104.csv')
# csv 파일로 저장할때 UTF-8 인코딩으로 저장하는것이 default
```

```python
with open('ns_202104.csv') as f:
    for i in range(3):
        print(f.readline(), end='')
# 이번에는 인덱스들도 출력된 모습
```

```python
ns_df = pd.read_csv('ns_202104.csv', low_memory=False)
ns_df.head()
# 인덱스가 다시 생성되어 버림
# csv파일에 인덱스가 이미 있다는 것을 알려줘야함
```

```python
ns_df = pd.read_csv('ns_202104.csv', index_col=0, low_memory=False)
ns_df.head()
# 'ns_202104.csv'에는 첫번째 열에 인덱스가 있으므로 index_col을 0으로 지정
```
