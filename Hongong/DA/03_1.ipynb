```python
# columns 속성은 Index 클래스 객체 -> 이름말고 숫자 인덱스로도 참조가능
print(ns_df.columns[0])
```

```python
# Index 클래스 같은 판다스 배열 성격의 객체는 어떤 값과 비교할 때
# 자동으로 배열에 있는 모든 원소와 하나씩 비교 해줌 -> 원소별 비교
ns_df.columns != 'Unnamed: 13'
# 원소별 비교 연산을 통해 불리언 배열을 얻음(넘파이 배열 형식)
```

```python
selected_columns = ns_df.columns != 'Unnamed: 13'
# True인 열의 행만 선택
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

## drop()을 통해 열 삭제

```python
# drop(삭제하려는 열 or 행, axis=1)
# axis=1은 열 삭제, axis=0은 행 삭제(default)
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
# 리스트 형식으로 여러 열 삭제
ns_book = ns_df.drop(['부가기호','Unnamed: 13'], axis=1)
ns_book.head()
```

```python
# inplace=True 설정하여 선택한 데이터프레임에 drop() 적용한 결과를 덮어쓰기
# ns_df에 drop() 결과를 ns_book이라는 새로운 변수에 저장하는 것과 같은 효과
# 새로운 변수를 만들지 않아도 되는 방법
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

## dropna() : NaN이 하나라도 포함되어 있는 열이나 행 삭제

```python
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
# 모든 값이 NaN인 열이나 행만 삭제 -> how 매개변수를 all로 지정
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

```python
# [] 연산자와 슬라이싱을 이용해 행 삭제
ns_book2 = ns_book[2:] # 인덱스 '2부터 끝까지'
ns_book2.head()
```

```python
ns_book2 = ns_book[0:2]
# [] 연산자에 슬라이싱을 사용하면 loc 메서드에 사용하는 것과 다르게
# 파이썬 슬라이싱처럼 마지막 인덱스를 포함하지 않음
ns_book2.head()
```

```python
ns_book2 = ns_book.loc[0:2]
# 슬라이싱 방법 비교
ns_book2.head()
```

```python
# [] 연산자와 불리언 배열 사용해 원하는 행만 선택
# 출판사가 '한빛미디어'인 행만 선택
selected_rows = ns_df['출판사'] == '한빛미디어'
selected_rows
```

```python
ns_book2 = ns_book[selected_rows]
ns_book2.head()
```

```python
# loc 메서드와 불리언 배열을 사용해 원하는 행 선택
ns_book2 = ns_book.loc[selected_rows]
# ns_book.loc[selected_rows, :] -> 전체 열 선택을 명시적으로 표현, 생략 가능함
ns_book2.head()
```

## duplicated() : 중복된 행 찾기

```python
# duplicated()
# 중복된 행 중에서 처음 행을 제외한 나머지 행은 True
# 그 외 중복되지 않은 나머지 모든 행은 False로 표시하는 불리언 배열 반환
# --> True인 행만 제거 하면 됨
# 파이썬 내장함수 sum()을 이용하면 True를 1로 인식 -> 중복된 행 개수 찾기
sum(ns_book.duplicated())
```

```python
# ns_book df는 '번호' 열에 고유한 값을 가지고 있어서 중복 행이 나올 수 없음
# 하지만 특정 열을 기준으로 중복된 행은 찾을 수 있음
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head(20)
# '도서명', '저자', 'ISBN'이 모두 동일한 도서들이 출력됨

# '보건교사 안은영'은 같은 도서인데 두 권이 등록되어 있는 것으로 확인
# head(20)으로 일부분만 확인한 결과이므로, 중복 행 전체를 확인하면 더 있을수도 있음

# '파친코'도 역시 여러 권이 등록되있고, 권 칼럼에 1, 2라는 값이 있는것으로 보아
# '파친코 1', '파친코 2'처럼 시리즈로 제작된 도서인 것으로 확인
```

## groupby() : 그룹별로 모으기

```python
# 같은 도서의 대출건수는 합쳐야 함
# 단, 권 수가 한 개 이상인 도서는 서로 다른 책으로 간주 (파친코 1, 파친코 2)
# groupby() 메서드는 행을 합칠 때 기준이 되는 열을 지정함
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
count_df.head()
```

```python
# groupby에서 by= 를 통해 그룹으로 묶을 기준 열을 지정

# groupby는 기본적으로 기준 열에 NaN이 포함 되어 있으면 해당 행을 삭제함
# 해당 행이 삭제되면 대출건수에서 빠지기 때문에 dropna=False 를 통해 방지
# dropna=False는 연산할 때 NaN이 있는 행도 포함하겠다는 의미

# 정수타입인 '대출건수'를 합치기 위해서 sum()를 이용
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False)
loan_count = group_df.sum()
```

```python
# 위 처럼 groupby와 sum을 나누어서 호출해도 되지만, 한번에 처리하면 간결
loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
```

```python
loan_count.head()
# 기준 열이 굵은 글씨로 표시됨
# --> '도서명'에서 '권'까지가 기준 인덱스 열이고,
# 각 책의 대출건수를 더한 결과가 '대출건수' 열에 저장되었음을 알수 있음
```

## 원본 데이터 업데이트하기

```python
# 원본 데이터프레임 중복 행을 True 반환하는 불리언 배열 생성
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
# ~ 연산자를 사용해 중복 행을 False, 고유한 행을 True로 표시하는 불리언 배열 생성
unique_rows = ~dup_rows
# 원본 데이터프레임에 고유한 행만 선택한 데이터프레임을 덮어쓰기
ns_book3 = ns_book[unique_rows].copy()
ns_book3.head()
```

```python
# 중복 행 없는지 한번더 확인
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))
```

```python
# 인덱스 열 설정하기
ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
ns_book3.head()
# loan_count df와 똑같이 '도서명' ~ '권' 을 인덱스 열로 설정
```

## update() : 다른 df로 원본 df 값을 업데이트

```python
ns_book3.update(loan_count)
# 대출건수 높은 순서대로 정렬해서 업데이트가 잘되었는지 확인
ns_book3_sorted = ns_book3.sort_values(by='대출건수', ascending=False)
ns_book3_sorted.head()
```

```python
# 업데이트 잘 되었으면 인덱스 열을 해제
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
# 대출건수가 잘 합쳐졌는지 확인
sum(ns_book['대출건수']>100)
```

```python
sum(ns_book4['대출건수']>100)
# 대출건수 100회 이상인 책이 많이 늘었음 -> 중복 도서 대출건수를 합쳤기 때문
```

```python
# 인덱스 열 설정과 해제를 하면서 원본의 열 순서와 달라졌음
# 열 순서를 원래대로 재조정 -> 원본 칼럼을 전달
ns_book4 = ns_book4[ns_book.columns]
ns_book4.head()
```

```python
# 완성된 자료 저장
ns_book4.to_csv('ns_book4.csv', index=False)
```

# 앞선 과정을 일괄 처리하는 함수 만들기

```python
def data_cleaning(filename):
    """
    남산 도서관 장서 CSV 데이터 전처리 함수

    : param filename: CSV 파일 이름
    """
    import pandas as pd
    # 파일을 데이터프레임으로 읽습니다.
    ns_df = pd.read_csv(filename, low_memory=False)
    # 모든 값이 NaN인 열을 삭제합니다.
    ns_book = ns_df.dropna(axis=1, how='all')
    # 대출건수를 합치기 위해 필요한 행만 추출하여 count_df 데이터프레임을 만듭니다.
    count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
    # 도서명, 저자, ISBN, 권을 기준으로 대출건수를 groupby합니다.
    loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
    # 원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사합니다.
    dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
    unique_rows = ~dup_rows
    ns_book3 = ns_book[unique_rows].copy()
    # 도서명, 저자, ISBN, 권을 인덱스로 설정합니다.
    ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
    # load_count에 저장된 누적 대출건수를 업데이트합니다.
    ns_book3.update(loan_count)
    # 인덱스를 재설정합니다.
    ns_book4 = ns_book3.reset_index()
    # 원본 데이터프레임의 열 순서로 변경합니다.
    ns_book4 = ns_book4[ns_book.columns]

    return ns_book4
```

```python
# 함수가 정상적으로 작동되는지 테스트
new_ns_book4 = data_cleaning('ns_202104.csv')

ns_book4.equals(new_ns_book4)
# True -> 두 데이터프레임이 완전히 동일
# data_cleaning() 함수가 앞서 했던 일련의 과정들을 동일하게 수행함
```
