## 정규 표현식

```python
# 모든 발행년도를 2자리로 표현하고 싶지만, 아래 코드는 '2021' -> '21'로 바꿈
ns_book4.replace({'발행년도': {'2021': '21'}})[100:102]
# replace()[] 처럼 뒤에 대괄호가 붙어도 되는 이유 -> 데이터프레임이기 때문
# 위의 식은 아래의 코드 두줄을 실행한것과 같은 결과

# ns_temp = ns_book4.replace({'발행년도': {'2021': '21'}})
# ns_temp[100:102].head(2)
```

```python
# 정규 표현식(정규식)을 사용하여 발행년도를 일괄적으로 끝 2자리만 남김
ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]
# \d -> 숫자를 찾는 기호
# \d\d\d\d는 4자리 숫자를 의미하는 패턴, ()를 이용해 그룹화 가능 -> \d\d(\d\d)
# 패턴 속에 그룹이 여러개 있어도 그룹마다 번호가 매겨짐 -> \1, \2 (번호는 패턴안에 먼저 등장하는 순서)
# regex=True: 정규 표현식을 사용한다는 의미
# 'r' 문자는 파이썬에서 정규 표현식과 일반 문자열을 구분하기 위해 붙이는 접두사
# 정규 표현식을 사용할 때 일반 문자열과 같이 사용하지 않으면 'r'을 붙이지 않아도 정규 표현식으로 인식함
```

```python
# 정규 표현식이 반복될 때는 중괄호를 사용해 개수를 지정
# \d{3} == \d\d\d
# \d\d(\d\d) == \d{2}(\d{2})
ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
ns_book4.replace({'발행년도': r'\d{2}(\d{2})'}, r'\1', regex=True)[100:102]
# 2코드 동일한 문장임 (참고)
```

```python
ns_book4.replace({'저자': {r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)': r'\1\2'},
                  '발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
# . : 문자열을 찾는 기호
# 발행년도는 4자리수 숫자임을 알기 때문에 \d{4}로 지정할수 있지만
# 저자는 사람 이름이므로 몇자리 문자열인지 미리 알지 못함
# .* : '0개 이상'을 의미하는 *를 이용, '임의의 길이의 문자열'을 뜻함 -> ()를 씌워 그룹핑
# \s : 공백을 찾는 기호
# \(지은이, \) -> \(지은이\) : '(지은이)' 라는 문자열로 인식하도록 표현
# \(지은이) 라고 해버리면 '지은이'를 그룹핑 했다고 인식하여 오류 유발
# \1\2는 그룹핑한 2개의 그룹을 반환 -> 2개의 이름을 반환하게 됨(지은이와 옮긴이의 이름 각각)
```

```python
# 뷰티풀수프 활용하여 누락정보 채우기
import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(isbn))
    # url주소에 {}을 포함시켜 .format() 형식의 문자열 포멧팅을 진행
    # 함수가 isbn 번호를 입력값으로 받으면 isbn을 검색한 페이지를 requests.get()
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    # 클래스 이름이 'gd_name'인 a 태그의 텍스트를 가져옵니다.
    title = soup.find('a', attrs={'class':'gd_name'}) \
            .get_text()
    return title
```

```python
# 도서명과 달리 저자는 두 명 이상수 있기 때문에 find_all()

import re

def get_book_info(row):
    # 데이터프레임의 '도서명' 칼럼에서 제목 정보를 추출해 title 변수에 저장
    title = row['도서명']
    author = row['저자']
    pub = row['출판사']
    year = row['발행년도']
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    # 데이터프레임의 'ISBN' 칼럼에서 추출한 도서번호를 이용해 HTML 가져오기
    # URL에 ISBN을 넣어 HTML 가져옵니다.
    r = requests.get(url.format(row['ISBN']))
    soup = BeautifulSoup(r.text, 'html.parser')   # HTML 파싱
    try:
        if pd.isna(title):
        # title의 값이 NaN 이면 -> 데이터프레임 '도서명' 칼럼에 결측치가 있다면
            # 클래스 이름이 'gd_name'인 a 태그의 텍스트를 가져옵니다.
            title = soup.find('a', attrs={'class':'gd_name'}) \
                    .get_text()
    except AttributeError:
        # AttributeError가 발생해도 코드가 정지되는 것이 아니라 그냥 넘어가라
        # AttributeError가 발생하는 상황 -> 클래스 이름이 'gd_name'인 a 태그가 없을 때
        # -> 태그가 없으면 soup.find()는 none을 반환하고 이것을 get.text() 하면 Error
        # 웹 페이지의 구조가 변경되거나, 요청한 정보가 없을때 발생하는 예외를 처리함
        pass

    try:
        if pd.isna(author):
            # 클래스 이름이 'info_auth'인 span 태그 아래 a 태그의 텍스트를 가져옵니다.
            authors = soup.find('span', attrs={'class':'info_auth'}) \
                          .find_all('a')
                          # 작가가 여러명 일수 있으므로, find_all()
            author_list = [auth.get_text() for auth in authors]
            # 리스트 컴프리헨션 구문 [expression for item in iterable if condition]
            # expression: 새로운 리스트의 각 요소가 될 표현식
            # item: iterable의 각 요소를 순회할 때 사용되는 변수입니다.
            # iterable: 리스트, 문자열, 범위(range) 등 반복 가능한 객체입니다.
            # if condition (optional): 조건이 참일 때만 expression을 평가하고
            #                          새로운 리스트에 포함시킵니다.
            # authors -> [<a>Author 1</a>, <a>Author 2</a>]
            # auth.get_text() -> Author 1
            # for auth in authors -> authors 리스트의 각 요소를 순회
            # 참고: 리스트의 각 요소도 인덱스로 추출할 수 있음. -> authors[0], authors[1]
            # author_list -> ['Author 1', 'Author 2']
            author = ','.join(author_list)
            # join() 메서드는 str 클래스의 메서드중 하나임 (문자열을 표현하는 그 클래스 맞음)
            # str 클래스가 파이썬 기본 내장 메서드 이므로 join() 메서드도 별도의 임포트 필요 없음
            # str 클래스에는 join() 말고도, upper(), lower() 등등 있음. str.join()
            # join() 메서드는 iterable 요소들을 특정 구분자로 연결하여 하나의 문자열로 만듦
            # 작가라는 칼럼에 데이터 타입이 문자열이 들어가야 하지, 리스트가 들어가면 안됨
            # [작가1, 작가2]라는 리스트를 '작가1, 작가2' 형식의 하나의 문자열로 만드는 과정
    except AttributeError:
        pass

    try:
        if pd.isna(pub):
            # 클래스 이름이 'info_auth'인 span 태그 아래 a 태그의 텍스트를 가져옵니다.
            pub = soup.find('span', attrs={'class':'info_pub'}) \
                      .find('a') \
                      .get_text()
    except AttributeError:
        pass

    try:
        if year == -1:
            # 클래스 이름이 'info_date'인 span 태그 아래 텍스트를 가져옵니다.
            year_str = soup.find('span', attrs={'class':'info_date'}) \
                           .get_text()
            # 정규식으로 찾은 값 중에 첫 번째 것만 사용합니다.
            year = re.findall(r'\d{4}', year_str)[0]
            # re 라이브러리는 파이썬 표준 라이브러리로 별도 설치 필요없음
            # math, datetime, os 등도 표준 라이브러리임
            # re 라이브러리는 정규 표현식을 다루기 위해 사용됨
            # re.findall(pattern, string, flags=0)
            # 정규 표현식과 일치하는 모든 부분 문자열을 리스트로 반환함
            # pattern: 검색할 정규 표현식
            # string: 검색할 문자열
            # flags: (선택 사항) 정규 표현식의 동작을 변경
            #        예를 들어 대소문자를 구분하지 않는 검색을 위해 re.IGNORECASE를 사용
    except AttributeError:
        pass

    return title, author, pub, year
```

```python
# axis=1: 행에 함수 적용 <-> axis=0: 열에 함수 적용
# result_type='expand': 함수의 결과가 여러 개면, 이를 개별 열로 확장
updated_sample = ns_book5[na_rows].head(2).apply(get_book_info,
    axis=1, result_type ='expand')
# 앞서 정의했듯, na_rows는 결측치 여부에 따른 불리언 시리즈임
# ns_book5[na_rows]는 ns_book5 데이터프레임의 각 행에 대해
# na_rows가 True인 행들만 선택해서 새로운 데이터프레임을 생성 -> 불리언 인덱싱의 개념
# ns_book5['도서명'] 처럼 df의 특정 열(칼럼)을 추출하는 개념과는 다름
# ns_book5[na_rows]는 df에서 'na_rows' 조건에 만족하는 행을 추출할 수 있는 편리한 방법임
updated_sample
```

```python
# 지금까지의 과정을 일괄 처리하는 함수
# 정규 표현식을 사용해 잘못된 값을 고치고, NaN(결측값)을 뷰티풀수프로 채우기

def data_fixing(ns_book4):
    """
    잘못된 값을 수정하거나 NaN 값을 채우는 함수

    :param ns_book4: data_cleaning() 함수에서 전처리된 데이터프레임
    """
    # 도서권수와 대출건수를 int32로 바꿉니다.
    ns_book4 = ns_book4.astype({'도서권수':'int32', '대출건수': 'int32'})
    # NaN인 세트 ISBN을 빈문자열로 바꿉니다.
    set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
    ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''

    # 발행년도 열에서 연도 네 자리를 추출하여 대체합니다. 나머지 발행년도는 -1로 바꿉니다.
    ns_book5 = ns_book4.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex=True)
    unkown_year = ns_book5['발행년도'].str.contains('\D', na=True)
    ns_book5.loc[unkown_year, '발행년도'] = '-1'

    # 발행년도를 int32로 바꿉니다.
    ns_book5 = ns_book5.astype({'발행년도': 'int32'})
    # 4000년 이상인 경우 2333년을 뺍니다.
    dangun_yy_rows = ns_book5['발행년도'].gt(4000)
    ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333
    # 여전히 4000년 이상인 경우 -1로 바꿉니다.
    dangun_year = ns_book5['발행년도'].gt(4000)
    ns_book5.loc[dangun_year, '발행년도'] = -1
    # 0~1900년 사이의 발행년도는 -1로 바꿉니다.
    old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
    ns_book5.loc[old_books, '발행년도'] = -1

    # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1인 행을 찾습니다.
    na_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna() \
              | ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)
    # 교보문고 도서 상세 페이지에서 누락된 정보를 채웁니다.
    updated_sample = ns_book5[na_rows].apply(get_book_info,
        axis=1, result_type ='expand')
    updated_sample.columns = ['도서명','저자','출판사','발행년도']
    ns_book5.update(updated_sample)

    # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1인 행을 삭제합니다.
    ns_book6 = ns_book5.dropna(subset=['도서명','저자','출판사'])
    ns_book6 = ns_book6[ns_book6['발행년도'] != -1]

    return ns_book6
```
