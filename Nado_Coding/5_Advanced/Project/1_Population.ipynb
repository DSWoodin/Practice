# 1. 인구 피라미드

## 2011년 데이터

### 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) # str 문자열(1,195,951)에서 ,를 공백으로 바꾼 뒤 정수형으로 반환
df_m
```

```python
df_m.iloc[0]
```

### 여자 데이터 정의

```python
df_w = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns
# 남자 데이터에도 '0~4세' 라는 column이 있기 때문에 구분하기 위해 자동으로 .1이 붙여짐
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w
```

### 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 남자 데이터에 -를 붙임으로써 0을 기준으로 남자는 왼쪽으로, 여자는 오른쪽으로
plt.title('2011 대한민국 인구 피라미드')
plt.xticks([-2000, -1000, 0, 1000, 2000], [2000, 1000, 0, 1000, 2000])
plt.savefig('2011_인구피라미드.png', dpi=100)
```

## 2021년 데이터

```python
df_m = pd.read_excel('202108_202108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:Y')
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_w = pd.read_excel('202108_202108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,AB:AV')
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w.columns = df_m.columns

plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 남자 데이터에 -를 붙임으로써 0을 기준으로 남자는 왼쪽으로, 여자는 오른쪽으로
plt.title('2021 대한민국 인구 피라미드')
plt.xticks([-2000, -1000, 0, 1000, 2000], [2000, 1000, 0, 1000, 2000])
plt.savefig('2021_인구피라미드.png', dpi=100)
```
