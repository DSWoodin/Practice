```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500 # 1학년=500, 2학년=1000, 3학년=1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

### matplotlib 사이트를 통해 cmap의 다양한 파라미터 확인가능
https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
sizes = df['학년'] * 500
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis')
# c= 각 점의 색상을 지정(숫자, 범주형 데이터, 색상값으로 지정)
# 여기서는 df['학년'] 이라는 범주형 데이터를 입력함
# cmap= 색상 맵을 설정함, viridis라는 color map은 값이 작으면 보라색, 크면 노란색에 가까워짐
# cmap='viridis'으로 설정했기 때문에 '1'학년은 보라색, 2학년 3학년으로 갈수록 노란색에 가까워짐
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
```
