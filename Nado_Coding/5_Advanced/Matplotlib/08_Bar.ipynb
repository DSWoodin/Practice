# 8. 막대 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False 
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values) # plt.bar(x값, y값)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195) # ylimit -> y축의 데이터를 어디서부터 어디까지 보여줄지
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y 축의 키 데이터 각도를 45도로 설정
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90) # x축의 이름 데이터 값을 변경
```

### matplotlib 사이트를 통해 hatch의 다양한 파라미터 확인가능
https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ....
# bar[0] -> 강백호 데이터
# bar[1] -> 서태웅 데이터
# bar[2] -> 정대만 데이터
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar): # rect 에는 각 막대 그래프의 속성에 대한 정보가 들어감
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='blue') 
    # rect.get_height() -> 각 막대 그래프의 높이 정보
    # plt.text(x, y, s, ha, color) -> plt.text(텍스트 x 위치, 텍스트 y 위치, 텍스트 내용, ha, color)
```
