# 능동 외각선
1987년 Kass는 능동 외각선(active contour) 알고리즘을 발표, 이는 사용자가 물체 내부에 초기 곡선을 지정하면 곡선을 점점 확장하면서 물체 외각선으로 접근한다.

곡선이 꿈틀대면서 외각선으로 가기 때문에 스네이크(Snake)라고 불리며, 스네이크 모델은 에너지 최소화 원리에 기초한다.

스네이크를 구현하려면 곡선 표현 방법이 필요한데, 이는 스플라인 곡선(spline curve)를 사용한다.

### 디지털 공간에서의 스네이크 곡선 표현

스플라인 곡선 $g(l)$은 2차원 상의 곡선이므로  $g(l)=(y(l),x(l))$로 표현한다.

매개변수 $l$은 $[0,1]$ 범위의 실수인데, 폐곡선을 이루기 위해 $g(0)=g(1)$이다.

디지털 영상은 이산 공간이기 때문에 $l$을 $0, 1, 2, ..., n$으로 표현한다.

스네이크 곡선 $g$의 에너지 $E$는 영상 에너지 $e_{image}$, 내부 에너지 $e_{internal}$, 도메인 에너지 $e_{domain}$의 합으로 표현한다.

> $E(g) = \sum_{l=0}^{n}(e_{image}(g(l))+e_{internal}(g(l))+e_{domain}(g(l)))$

영상 에너지 $e_{image}$는 물체 경계에 에지가 나타난다는 사실에 기반하여 곡선이 에지에 위치하도록 유도한다.

내부 에너지 $e_{internal}$는 일반적으로 물체 경계가 급격하게 꺽이는 곳보다 매끄럽게 변하는 곳이 더 많다는 사실에 기반하여 곡선이 매끄러운 모양이 되도록 유도한다.

도메인 에너지 $e_{domain}$는 분할하려는 특정 물체의 모양 정보를 잘 유지하도록 유도한다.

실제 구현에서는 영상 에너지 $e_{image}$를 위해 에지 강도를 사용하고 내부 에너지 $e_{internal}$를 위해 곡률을 사용한다.

앞의 두 가지 에너지는 물체의 일반 특성에 관련되어 사용자가 개입하지 않아도 동작하지만, 도메인 에너지 $e_{domain}$는 사용자가 모양을 지정해야 하는데 종종 무시한다.

아래의 식은 스네이크 알고리즘이 풀어야 하는 문제를 정의하는데, 에너지를 최소로 하는 최적 곡선을 찾는 최적화 문제이다. 어떻게 최적 곡선을 빨리 찾을 수 있을까?

> $\hat{g} = argminE(g)$  --> 함수 $E(g)$를 최소화하는 최적의 $g$값, 즉 $\hat{g}$를 찾겠다

스네이크 알고리즘은 사용자가 지정해준 초기 곡선을 $g_{0}$으로 두고 $g_{0}$에서 $g_{1}$, $g_{1}$에서 $g_{2}$, ..., $g_{t}$에서 $g_{t+1}$을 찾아가는 과정을 수렴할 때까지 반복한다.

# GrabCut

네트워크 흐름(network flow)은 여러 지점을 거치는 물이나 전기의 흐름을 그래프로 표현하고 병목 지점을 찾아 흐름을 개선하는 문제로, 오래 전부터 연구되어왔다.

Greig는 네트워크 흐름 알고리즘을 영상 복원에 활용하는 발상을 제시하여 컴퓨터 비전 분야로 끌어들였다.

한동안 별로 주목을 받지 못하다가 10여 년이 지나 스테레오와 영역 분할 등의 여러 문제에 활발히 적용되기 시작했다.

현재까지 널리 쓰이는 연구 결과는 GrabCut 알고리즘이다.

1. GrabCut에서는 사용자가 붓으로 물체와 배경을 초기 지정한다.

2. 화소를 이용해 물체 히스토그램과 배경 히스토그램을 만든다.

3. 나머지 화소들은 두 히스토그램과 유사성을 따져 물체일 확률과 배경일 확률을 추정하고 물체 영역과 배경 영역을 갱신한다.

4. 2~3번 과정을 반복하여 영역이 거의 변하지 않으면 수렴한 것으로 간주하고 알고리즘이 종료된다.
